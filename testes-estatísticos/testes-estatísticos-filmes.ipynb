```python
import pandas as pd
import seaborn as sns
```

```python
tmdb = pd.read_csv("tmdb_5000_movies.csv")
tmdb.head()
```

```python
tmdb.describe()
```

```python
ax = sns.distplot(tmdb.vote_average)
ax.set(xlabel = "Notas", ylabel = "Densidade")
ax.set_title("Média dos filmes no TMDB 5000")
```

```python
ax = sns.distplot(tmdb.vote_average, norm_hist=False, kde=False)
ax.set(xlabel = "Notas", ylabel = "Frequência")
ax.set_title("Média dos filmes no TMDB 5000")
```

```python
ax = sns.boxplot(x= tmdb.vote_average)
ax.set_title("Distribuição das notas")
ax.set(xlabel='Notas')
```

```python
tmdb_com_mais_10votos = tmdb.query("vote_count >= 10")
tmdb_com_mais_10votos.head()
```

```python
ax = sns.distplot(tmdb_com_mais_10votos.vote_average, norm_hist=False, kde=False)
ax.set(xlabel = "Notas", ylabel = "Frequência")
ax.set_title("Média dos filmes no TMDB 5000 com mais de 10 votos")
```

```python
ax = sns.boxplot(x= tmdb_com_mais_10votos.vote_average)
ax.set_title("Distribuição das notas")
ax.set(xlabel='Notas')
```

```python
notas = pd.read_csv("ratings.csv")
notas.head()
```

```python
nota_media_por_filme = notas.groupby("movieId").mean()["rating"]
```

```python
ax = sns.distplot(nota_media_por_filme.values)
ax.set(xlabel = "Notas", ylabel = "Frequência")
ax.set_title("Média dos filmes no TMDB 5000 com mais de 10 votos")
```

```python
quantidade_de_votos_por_filme = notas.groupby("movieId").count()
filmes_com_poelo_menos_10_votos = quantidade_de_votos_por_filme.query("rating >= 10").index
```

```python
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_poelo_menos_10_votos.values]
```

```python
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values)
ax.set(xlabel = "Notas", ylabel = "Frequência")
ax.set_title("Média dos filmes no TMDB 5000 com mais de 10 votos")
```

```python
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values, hist_kws={'cumulative':True}, kde={'cumulative':True})
ax.set(xlabel = "Notas", ylabel = "% acumuldade de filmes")
ax.set_title("Média dos filmes no TMDB 5000 com mais de 10 votos")
```

```python
import matplotlib.pyplot as plt
import numpy as np

medias = list()
for i in range(1, len(nota_media_dos_filmes_com_pelo_menos_10_votos)):
    medias.append(nota_media_dos_filmes_com_pelo_menos_10_votos[0:i].mean())

plt.plot(medias)
```

```python
temp = nota_media_dos_filmes_com_pelo_menos_10_votos.sample(frac=1)

medias = list()
medias = [temp[0:i].mean() for i in range(1, len(temp))]

plt.plot(medias)
```
