```python
notas = pd.read_csv("ratings.csv")
notas.head()
```

```python
nota_media_por_filme = notas.groupby("movieId").mean()["rating"]
```

```python
ax = sns.distplot(nota_media_por_filme.values)
ax.set(xlabel = "Notas", ylabel = "Frequência")
ax.set_title("Média dos filmes no TMDB 5000 com mais de 10 votos")
```

```python
quantidade_de_votos_por_filme = notas.groupby("movieId").count()
filmes_com_poelo_menos_10_votos = quantidade_de_votos_por_filme.query("rating >= 10").index
```

```python
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_poelo_menos_10_votos.values]
```

```python
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values)
ax.set(xlabel = "Notas", ylabel = "Frequência")
ax.set_title("Média dos filmes no TMDB 5000 com mais de 10 votos")
```

```python
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values, hist_kws={'cumulative':True}, kde={'cumulative':True})
ax.set(xlabel = "Notas", ylabel = "% acumuldade de filmes")
ax.set_title("Média dos filmes no TMDB 5000 com mais de 10 votos")
```

```python
from statsmodels.stats.weightstats import ztest
ztest(notas1.rating, value = 3.4320)
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i], value = 3.4320)
    return (i, media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])

plt.plot(valores[ : , 0], valores[ : , 1])
plt.plot(valores[: , 0], valores[ : , 2])
plt.hlines(y = 0.05, xmin=2, xmax = len(temp), colors='r')
```
